```python
import pandas as pd

msg = pd.read_csv(
    "C:/Users/prath/Downloads/spam.csv",
 
    encoding="ISO-8859-1"   # Or encoding="latin1"
)

print(msg.head())
```

```python
msg = msg.drop(columns=["Unnamed: 2" ,'Unnamed: 3',"Unnamed: 4"],axis=1)
```

```python
msg
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
ps =PorterStemmer()
```

```python
corpus = []                     ## Time taking process
for i in range(0,len(msg)):
    review = re.sub('[^a-zA-Z]',' ' ,msg['v2'][i])
    review = review.lower()
    review = review.split()
    review =[ps.stem(word.lower()) for word in review if not word.lower() in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

## create a Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True)     ## IMP param i.e. max_features,also binary get only 0 and 1
```

```python
x= cv.fit_transform(corpus).toarray()
```

```python
x
```

### n_grams

```python
## Unigrams

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100,binary=True,ngram_range=(1,1))    ## (1,1) for unigrams
```

```python
x= cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```
